```python
from gensim.models import KeyedVectors
```

```python
filename = 'GoogleNews-vectors-negative300.bin'
print("test")
model = KeyedVectors.load_word2vec_format(filename, binary=True)
```

```python
similar_words = model.most_similar('kawaii')
print(similar_words)
```

```python
if 'lotus' in model:
    vector_lotus = model['lotus']
    vector_flower = model['flower']
    
    print("Vector representation of 'lotus':")
    print(vector_lotus)
    
    print("\nVector representation of 'flower':")
    print(vector_flower)
```

```python
word1 = 'lotus'
word2 = 'flower'

if word1 in model and word2 in model:
    similarity_score = model.similarity(word1, word2)
    print(f"Similarity between '{word1}' and '{word2}': {similarity_score:.4f}")
else:
    print(f"One or both of the words '{word1}' and '{word2}' are not in the vocabulary.")
```

```python
word3 = 'white'
similarity_score = model.similarity(word1, word3)
print(f"Similarity between '{word1}' and '{word3}': {similarity_score:.4f}")

word31 = (model[word3] + model[word1]) / 2.0
word32 = (model[word3] + model[word2]) / 2.0
```

```python
# Compute similarity between word32 and word1
similarity_score = model.cosine_similarities(model[word1], [word31])[0]
print(f"Similarity between '{word1}' and '{word3 + word1}': {similarity_score:.4f}")
```

```python
word4 = 'pretty'
word5 = 'flower'
word6 = 'dead'
word7 = 'lily'
```

```python
similarity_score = model.similarity(word4, word5)
print(f"Similarity between '{word4}' and '{word5}': {similarity_score:.4f}")

similarity_score = model.similarity(word4, word6)
print(f"Similarity between '{word4}' and '{word6}': {similarity_score:.4f}")

similarity_score = model.similarity(word4, word7)
print(f"Similarity between '{word4}' and '{word7}': {similarity_score:.4f}")

similarity_score = model.similarity(word6, word5)
print(f"Similarity between '{word6}' and '{word5}': {similarity_score:.4f}")

similarity_score = model.similarity(word6, word7)
print(f"Similarity between '{word6}' and '{word7}': {similarity_score:.4f}")

similarity_score = model.similarity(word7, word5)
print(f"Similarity between '{word7}' and '{word5}': {similarity_score:.4f}")
```

```python
pretty_flower = (model[word4] + model[word5]) / 2.0
pretty_lily = (model[word4] + model[word7]) / 2.0
dead_flower = (model[word6] + model[word5]) / 2.0
dead_lily = (model[word6] + model[word7]) / 2.0
```

```python
# Compute similarity 
similarity_score = model.cosine_similarities([pretty_flower][0], [pretty_lily])[0]
print(f"Similarity between pretty_lily and pretty_flower': {similarity_score:.4f}")

similarity_score = model.cosine_similarities([pretty_flower][0], [dead_lily])[0]
print(f"Similarity between dead_lily and pretty_flower': {similarity_score:.4f}")

similarity_score = model.cosine_similarities([dead_flower][0], [pretty_lily])[0]
print(f"Similarity between pretty_lily and dead_flower': {similarity_score:.4f}")

similarity_score = model.cosine_similarities([dead_flower][0], [dead_lily])[0]
print(f"Similarity between dead_lily and dead_flower': {similarity_score:.4f}")
```
